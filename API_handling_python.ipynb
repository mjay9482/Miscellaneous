```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://api.themoviedb.org/3/movie/top_rated?api_key=71f016afd6138629dd65339227ac3f16&language=en-US&page=1"
response = requests.get(url)
response.json()
```

```python
df = pd.DataFrame(response.json()['results'])[['id', 'title','overview', 'release_date','popularity','vote_average','vote_count']]
```

```python
df.head()
```

```python
dfs = []
for i in range(1, 474):
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=71f016afd6138629dd65339227ac3f16&language=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title','overview', 'release_date','popularity','vote_average','vote_count']]
    dfs.append(temp_df)
df = pd.concat(dfs, ignore_index=True)
```
